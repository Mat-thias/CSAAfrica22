```python
class KNN:
    """An implementation of k nearest neighbour"""
    def __init__(self, k_neighbors):
        self.k_neighbors = k_neighbors
        
    def fit(self,  X_train, y_train):
        if X_train.shape[0] != y_train.shape[0]:
            raise Exception("In compatable shapes")
        self.X_train = X_train
        self.y_train = y_train
    
    def predict_point(self, x):
        dist_arr = ((self.X_train - x) ** 2).sum(axis=1)
        sort_dist_y = sorted(zip(dist_arr, self.y_train), key=lambda x : x[0])
        nearest_n = np.array(sort_dist_y[:self.k_neighbors])[:,1]
        
        nearest_n_dict = {}
        for y in nearest_n:
            nearest_n_dict[y] = nearest_n_dict.get(y, 0) + 1
        
        sort_nearest_n_dist = sorted(nearest_n_dict.items(), key=lambda x : x[1])
        
        return sort_nearest_n_dist[0][0]
    
    def predict(self, testX):
        return_list = []
        for x in testX:
            return_list.append(self.predict_point(x))
            
        return np.array(return_list)
    
    def accuracy(self, testX, testy):
        if len(testX) != len(testy):
            return "Incompatable sizes."
        
        pred_y = self.predict(testX)
        acc_list = [] 
        for i in range(len(testy)):
            acc_list.append(testy[i] == pred_y[i])
            
        return acc_list.count(True) / len(acc_list)
```

```python
class KNNDist:
    """An implementation of k nearest neighbour with dist"""
    def __init__(self, k_neighbors):
        self.k_neighbors = k_neighbors
        
    def fit(self,  X_train, y_train):
        if X_train.shape[0] != y_train.shape[0]:
            raise Exception("In compatable shapes")
        self.X_train = X_train
        self.y_train = y_train
        
    def predict_point(self, x):
        dist_arr = ((self.X_train - x) ** 2).sum(axis=1)
        sort_dist_y = sorted(zip(dist_arr, self.y_train), key=lambda x : x[0])
        nearest_n = np.array(sort_dist_y[:self.k_neighbors])
        
        nearest_n_dict = {}
        for dist_y  in nearest_n:
            dist, y = dist_y
            nearest_n_dict[y] = nearest_n_dict.get(y, 0) + dist**2
        
        sort_nearest_n_dist = sorted(nearest_n_dict.items(), key=lambda x : x[1])
        
        return sort_nearest_n_dist[0][0]
    
    def predict(self, testX):
        return_list = []
        for x in testX:
            return_list.append(self.predict_point(x))
            
        return np.array(return_list)
    
    def accuracy(self, testX, testy):
        if len(testX) != len(testy):
            raise Exception("Incompatable sizes.")
        
        pred_y = self.predict(testX)
        acc_list = [] 
        for i in range(len(testy)):
            acc_list.append(testy[i] == pred_y[i])
            
        return acc_list.count(True) / len(acc_list)
```

```python
def compare_KNN_with_sklearn_KClassifier(X_train, X_test, y_train, y_test, k_range=list(range(1,100))):
    k_len = len(k_range)

    acc_array = []
    for k in k_range:
        cancer_neigh = KNeighborsClassifier(n_neighbors=k)
        cancer_neigh.fit(X_train, y_train)
        acc_array.append(cancer_neigh.score(X_test, y_test))

    plt.plot(k_range, acc_array, label="sklearn")
    plt.xlabel("k")
    plt.ylabel("accuracy")

    acc_array = []
    for k in k_range:
        knn = KNN(k_neighbors=k)
        knn.fit(X_train, y_train)
        acc_array.append(knn.accuracy(X_test, y_test))

    plt.plot(k_range, acc_array, label="Matthias")
    plt.xlabel("k")
    plt.ylabel("accuracy")

    acc_array = []
    for k in k_range:
        knn = KNNDist(k_neighbors=k)
        knn.fit(X_train, y_train)
        acc_array.append(knn.accuracy(X_test, y_test))

    plt.plot(k_range, acc_array, label="dist")
    plt.xlabel("k")
    plt.ylabel("accuracy")


    plt.legend()
    plt.show()
```
